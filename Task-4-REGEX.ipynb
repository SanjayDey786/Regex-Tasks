# 3. The following columns need to be converted for further analysis:

## 'Value'
## #Convert to Float after getting rid of currency symbol and suffix.

```python
df_fifa['Value']
```

```python
df_fifa['Value'] = df_fifa['Value'].str.replace('€','')
```

```python
def value_con(value):
    if 'K' in value:
        value = value.replace('K','')
        value = float(value)/1000
        return str(value)
    else:
        if 'M' in value:
            value=value.replace('M','')
            return value
```

```python
df_fifa['Value'] = df_fifa['Value'].apply(value_con)
```

```python
df_fifa['Value'] = df_fifa['Value'].astype('float')
```

```python
df_fifa['Value'].dtype
```
